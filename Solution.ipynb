## Imports

```python
import pandas as pd

from pandas.tseries.offsets import BDay
```

## Define functions and load data

```python
def read_file(filename):
    df = pd.read_csv(filename, header=0, encoding='gb2312')
    
    # rename columns for better indexing/sorting
    df.columns = [
        "mkt_code",
        "contract_code",
        "time",
        "latest",
        "hold",
        "add",
        "turnover",
        "volume",
        "kaicang",
        "pingcang",
        "turnover_type",
        "direction",
        "buy_first_price",
        "sell_first_price",
        "buy_first_volume",
        "sell_first_volume",
        
    ]
    
    return df

def read_data():
    grand_df = None
    for i in range(8, 23):
        if (i != 12 and i != 13 and i != 19 and i != 20):
            filename = "./rb1605_201603" + (("0" + str(i)) if i < 10 else str(i)) + ".csv"

            df = read_file(filename)

            if i == 8:
                grand_df = df
            else:
                grand_df = grand_df.append(df)
        
    return grand_df

df = read_data()

# use merge sort to save time potentially
df = df.sort_values(by="time", kind="mergesort")
```

## Solve question 2)

```python
# reset index
df = df.set_index(pd.DatetimeIndex(pd.to_datetime(df["time"])))
df.index = pd.to_datetime(df.index, unit='ms');

# resample to 500ms
df.resample("500ms").sum()

# fill NaN value
df.fillna(method="ffill")

# remove weekends
df = df[df.index.dayofweek < 5]
```

```python
# get 9:00 - 11:30
df_9 = df[df.index.hour >= 9]
df_10 = df_9[df_9.index.hour <= 10]
df_11 = df[df.index.hour == 11]
df_1130 = df_11[df_11.index.minute <= 30]

df_9_to_1130 = df_10.append(df_1130)

# get 13:30 - 15:00
df_14 = df[df.index.hour >= 14]
df_15 = df_14[df_14.index.hour <= 15]
df_13 = df[df.index.hour == 13]
df_1330 = df_13[df_13.index.minute >= 30]

df_1330_to_15 = df_1330.append(df_15)

# get 21:00 - 1:00
df_1 = df[df.index.hour <= 1]
df_21 = df[df.index.hour >=21]

df_21_to_1 = df_1.append(df_21)

# merge all
final_df = df_9_to_1130.append(df_1330_to_15).append(df_21_to_1)

final_df
```
